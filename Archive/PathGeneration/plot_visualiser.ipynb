```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets
from IPython.display import display

# Define your function
def line(t, g, f, e, d, c, b, a, r, q, p, o, n, m, l):
    return (t, g*t**6 + f*t**5 + e*t**4 + d*t**3 + c*t**2 + b*t + a, r*t**6 + q*t**5 + p*t**4 + o*t**3 + n*t**2 + m*t + l)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create sliders for the coefficients
g_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='g:')
f_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='f:')
e_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='e:')
d_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='d:')
c_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='c:')
b_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='b:')
a_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='a:')

r_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='r:')
q_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='q:')
p_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='p:')
o_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='o:')
n_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='n:')
m_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='m:')
l_slider = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1, description='l:')

# Display the sliders
display(g_slider, f_slider, e_slider, d_slider, c_slider, b_slider, a_slider, r_slider, q_slider, p_slider, o_slider, n_slider, m_slider, l_slider)

# Define a function to update the plot
def update_plot(change):
    ax.clear()
    t = np.linspace(-10, 10, 100)
    x, y, z = line(t, g_slider.value, f_slider.value, e_slider.value, d_slider.value, c_slider.value, b_slider.value, a_slider.value, r_slider.value, q_slider.value, p_slider.value, o_slider.value, n_slider.value, m_slider.value, l_slider.value)
    ax.plot(x, y, z)
    
    # Set the limits of the axes
    ax.set_xlim([-20, 20])
    ax.set_ylim([-20, 20])
    ax.set_zlim([-20, 20])
    
    display(fig)  # Use display instead of plt.show()
# ... (same as before)

# Connect the sliders to the update function
g_slider.observe(update_plot, names='value')
f_slider.observe(update_plot, names='value')
e_slider.observe(update_plot, names='value')
d_slider.observe(update_plot, names='value')
c_slider.observe(update_plot, names='value')
b_slider.observe(update_plot, names='value')
a_slider.observe(update_plot, names='value')
r_slider.observe(update_plot, names='value')
q_slider.observe(update_plot, names='value')
p_slider.observe(update_plot, names='value')
o_slider.observe(update_plot, names='value')
n_slider.observe(update_plot, names='value')
m_slider.observe(update_plot, names='value')
l_slider.observe(update_plot, names='value')
```
